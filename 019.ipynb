 ## **As 10 regiões com maior interesse em "Machine Learning" de acordo com  o Google Trends**

# **OBJETIVO**
Este script em Python utiliza a API do Google Trends para analisar a popularidade da consulta "Machine Learning" em termos de interesse por região e ao longo do tempo.

**Detalhes do Código**
* A biblioteca pytrends é usada para acessar a API do Google Trends.
* Consultas são feitas para obter dados de interesse por região e ao longo do tempo.
* Pandas é utilizado para manipular e organizar os dados.
* Matplotlib é usado para visualizar os dados por região em um gráfico de barras e ao longo do tempo em um gráfico de séries temporais.

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["Machine Learning"])
data = trends.interest_by_region()
data = data.sort_values(by="Machine Learning", ascending=False)
data = data.head(10)
print(data)
```

**Os resultados obtidos representam as 10 regiões com maior interesse em "Machine Learning" de acordo com os dados fornecidos pelo Google Trends. O Google Trends avalia o interesse em uma determinada palavra-chave ao longo do tempo e em diferentes regiões geográficas. Os resultados são normalizados, onde o valor máximo para uma região específica é definido como 100, e outros valores são proporcionais a esse máximo.**

* O código define a palavra-chave como "Machine Learning" e prepara uma consulta ao Google Trends para obter dados relacionados a essa palavra-chave.

* Os dados de interesse regional são obtidos do Google Trends para a palavra-chave especificada. Cada região ou país tem um valor associado na coluna "Machine Learning", indicando o interesse relativo nessa região.

* Os dados são ordenados em ordem decrescente com base no interesse em "Machine Learning". As regiões com maior interesse ficam no topo da lista.

* São selecionadas as 10 primeiras regiões da lista ordenada, representando as
regiões com os maiores índices de interesse em "Machine Learning".

* A tabela resultante é impressa, mostrando o nome das regiões na coluna "geoName" e os valores normalizados de interesse em "Machine Learning" na coluna correspondente.

```python


data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data = data.interest_by_region(resolution='COUNTRY', inc_geo_code=False)

top_10_data = data.nlargest(10, 'Machine Learning')

top_10_data.reset_index().plot(x="geoName", y="Machine Learning", figsize=(10, 6), kind="bar")
plt.style.use('fivethirtyeight')
plt.title('Top 10 Regiões com Maior Interesse no Google para Aprendizado de Máquina', fontweight='bold')
plt.xlabel('Região')
plt.ylabel('Total Count')

plt.show()
```

```python
import matplotlib.pyplot as plt
from pytrends.request import TrendReq


data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data = data.interest_over_time()


fig, ax = plt.subplots(figsize=(10, 8))
data['Machine Learning'].plot()


plt.style.use('fivethirtyeight')
plt.title('Total de Pesquisas no Google para Aprendizado de Máquina',
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')


plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data = data.interest_over_time()


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))


ax1.plot(data.index, data['Machine Learning'], label='Machine Learning')
ax1.set_title('Pesquisas Google - Aprendizado de Máquina', fontweight='bold', fontsize=12)
ax1.set_xlabel('Ano')
ax1.set_ylabel('Contagem Total')
ax1.legend()


data['Mês'] = data.index.month
monthly_trend = data.groupby('Mês')['Machine Learning'].mean()

ax2.bar(monthly_trend.index, monthly_trend, color='skyblue')
ax2.set_title('Tendência Mensal - Aprendizado de Máquina', fontweight='bold', fontsize=12)
ax2.set_xlabel('Mês')
ax2.set_ylabel('Contagem Média')

plt.tight_layout()
plt.show()
```

A visualização por região é realizada por meio de um gráfico de barras, o que facilita a compreensão das áreas com maior interesse na consulta. A análise temporal é representada por um gráfico de séries temporais, proporcionando uma visão clara das variações no interesse ao longo dos anos.
