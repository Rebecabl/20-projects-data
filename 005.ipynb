```python
from google.colab import files
uploaded = files.upload()
import pandas as pd


file_name = list(uploaded.keys())[0]
population_df = pd.read_csv(file_name)

# Exibir as primeiras linhas do DataFrame para verificar se foi carregado corretamente
population_df.head()
```

```python
print(population_df.columns)
```

```python
print(population_df['Latest population census'].dtype)
```

```python
print(population_df['Latest population census'].isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Verificar dados ausentes
print("Dados ausentes:", population_df['Latest population census'].isnull().sum())

# Converter para tipo numérico, se possível
try:
    population_df['Latest population census'] = pd.to_numeric(population_df['Latest population census'])
except ValueError as e:
    print("Erro na conversão para tipo numérico:", e)

# Criar o histograma
plt.hist(population_df['Latest population census'].dropna(), bins=20, color='blue', alpha=0.7)
plt.xlabel('Último censo populacional')
plt.ylabel('Frequência')
plt.title('Histograma do Último Censo Populacional')
plt.show()
```
