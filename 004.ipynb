```python
import pandas as pd

# Substitua 'seu_arquivo.csv' pelo nome do arquivo que você fez o upload
file_name = list(uploaded.keys())[0]
population_df = pd.read_csv(file_name)

# Exibir as primeiras linhas do DataFrame para verificar se foi carregado corretamente
population_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))

# Gráfico de barras para Região
plt.subplot(2, 1, 1)
sns.countplot(x='Region', data=population_df, palette='viridis')
plt.title('Distribuição por Região')

# Gráfico de barras para Renda
plt.subplot(2, 1, 2)
sns.countplot(x='Income Group', data=population_df, palette='viridis')
plt.title('Distribuição por Grupo de Renda')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
region_counts = population_df['Region'].value_counts()
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis', len(region_counts)))
plt.title('Distribuição das Regiões')
plt.show()
```

```python
income_group_counts = population_df['Income Group'].value_counts()
plt.figure(figsize=(12, 8))
sns.countplot(x='Income Group', data=population_df, palette='viridis')
plt.title('Distribuição por Grupo de Renda')
plt.xlabel('Grupo de Renda')
plt.ylabel('Contagem')
plt.show()
```

```python

plt.figure(figsize=(14, 8))
stacked_bar_data = population_df['Income Group'].value_counts().reset_index()
sns.barplot(x='index', y='Income Group', data=stacked_bar_data, palette='viridis')
plt.title('Contagem de Valores Únicos por Grupo de Renda')
plt.xlabel('Grupo de Renda')
plt.ylabel('Contagem')
plt.show()
```
