# ***Análise de Dados dos Municípios Brasileiros***



### **Objetivo**

Este projeto tem como objetivo realizar uma análise exploratória dos dados dos municípios brasileiros. O conjunto de dados contém informações sobre 5.570 municípios, incluindo o código IBGE, nome do município, código da unidade federativa (UF), sigla da unidade federativa, nome do estado, indicador de capital, latitude e longitude.



#### **Etapas do Projeto**

*   Carregamento do Arquivo

*   Extração do Conteúdo do Arquivo ZIP

*   Carregamento dos Dados no DataFrame

*   Análise Exploratória dos Dados

*   Visualização dos Resultados

```python
import zipfile
import os
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Carregamento do Arquivo
# Carregar o arquivo zip

```python
from google.colab import files
uploaded = files.upload()
```

# O nome do arquivo zip enviado

```python
zip_file_path = list(uploaded.keys())[0]
extracted_dir_path = 'extracted'
```

# **2. Extração do Conteúdo do Arquivo ZIP**
# Extrair o conteúdo do arquivo zip

```python
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir_path)
```

# Listar os arquivos extraídos

```python
extracted_files = os.listdir(extracted_dir_path)
```

# Carregar o arquivo CSV em um DataFrame

```python
csv_file_path = os.path.join(extracted_dir_path, 'municipios_brasileiros.csv')
df = pd.read_csv(csv_file_path)
```

# Mostrar informações básicas e as primeiras linhas do DataFrame

```python
df.info()
df.head()
```

# **4. Análise Exploratória dos Dados**

## Contagem de Municípios por Estado
# Contagem de municípios por estado

```python
municipios_por_estado = df['estado'].value_counts()
```

# Plotar a distribuição de municípios por estado

```python
plt.figure(figsize=(12, 8))
municipios_por_estado.plot(kind='bar', color='skyblue')
plt.title('Distribuição dos Municípios por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Municípios')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Análise das Capitais
# Filtrar os municípios que são capitais

```python
capitais = df[df['capital'] == 1]
```

# Número de capitais

```python
numero_de_capitais = capitais.shape[0]
```

# Mostrar as capitais

```python
capitais_info = capitais[['nome_municipio', 'estado', 'latitude', 'longitude']]
print(f'Número de Capitais: {numero_de_capitais}')
print(capitais_info)
```

## Distribuição Geográfica
# Plotar o mapa do Brasil com os pontos dos municípios

```python
fig, ax = plt.subplots(figsize=(12, 12))
ax.scatter(df['longitude'], df['latitude'], s=1, color='blue', alpha=0.5)
ax.set_title('Distribuição Geográfica dos Municípios Brasileiros')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
plt.show()
```
